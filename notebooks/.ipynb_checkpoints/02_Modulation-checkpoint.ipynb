# 2. Modultaion

## Amplitude Modulation

```python
%pylab inline
from IPython.display import Audio
style.use('seaborn') #more beautiful plots
```

```python
fs = 44100.
T = 1.
N = int(T*fs)
n = arange(N)
t = n/fs
```

```python
f1 = 70
f2 = 1
x1 = sin(2*pi*t*f1)
x2 = sin(2*pi*t*f2)
y = x1*x2
```

```python
figure(figsize=[10,8])
subplot(311)
plot(t,x1)
subplot(312)
plot(t,x2)
subplot(313)
plot(t,y)
show()
```

```python
Audio(data=y, rate=fs)
```

## Frequency Modulation

```python
fs = 44100.
T = 10.
N = int(T*fs)
n = arange(N)
t = n/fs
```

```python
f1 = 1 #modulator
f2 = 150 #carrier
amnt = 100
x1 = sin(2*pi*t*f1)
carrierTerm = 2*pi*t*f2
fMod = x1*amnt
x2 = sin(carrierTerm+fMod)
```

```python
xlimit = [0,1/f1] #Limit the plot to 1 cycle of the modulator

figure(figsize=[10,8])
subplot(411)
plot(t,x1)
xlim(xlimit)
subplot(412)
plot(t,x1*amnt+f2)
xlim(xlimit)
subplot(413)
plot(t,x2)
xlim(xlimit)
subplot(414)
specgram(x2, Fs=fs, vmin=-150,NFFT=2048)
xlabel('time [sec]')
ylabel('Frequency [Hz]')
# colorbar()
xlim(xlimit)
ylim([clip(f2-amnt,20,2000),clip(f2+1000,20,2000)])
show()
```

```python
specgram(x2, Fs=fs, vmin=-150,NFFT=2048)
xlabel('time [sec]')
ylabel('Frequency [Hz]')
ylim([20,10000])
colorbar()
show()
```

```python
Audio(data=x2, rate=fs)
```
