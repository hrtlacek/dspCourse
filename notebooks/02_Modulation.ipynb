## Amplitude Modulation

```python
%pylab inline
from IPython.display import Audio
```

```python
fs = 44100.
T = 1.
N = int(T*fs)
n = arange(N)
t = n/fs
```

```python
f1 = 70
f2 = 1
x1 = sin(2*pi*t*f1)
x2 = sin(2*pi*t*f2)
y = x1*x2
```

```python
figure
subplot(311)
plot(t,x1)
subplot(312)
plot(t,x2)
subplot(313)
plot(t,y)
```

```python
Audio(data=y, rate=fs)
```

## Frequency Modulation

```python
fs = 600.
T = 10.
N = int(T*fs)
n = arange(N)
t = n/fs
```

```python
f1 = 1
f2 = 200
amnt = 2
x1 = sin(2*pi*t*f1)
fMod = f2+x1*amnt
x2 = sin(2*pi*fMod)
```

```python
figure
subplot(311)
plot(t,x1)
subplot(312)
plot(t,fMod)
subplot(313)
plot(t,x2)
show()
```

```python
specgram(x2, Fs=fs)
show()
```
